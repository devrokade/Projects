### Problem 2 (Bar chart)
a) Draw a vertical bar chart of the pclass variable with pclass in the x-axis and counts in the y-axis. Make a one-sentence comment about your plot. Read more about the bar chart here. https://dfrieds.com/data-visualizations/bar-plot-python-pandas.html

b) Draw a horizontal bar chart of the embarked variable with embarked labels on the y-axis and counts on the x-axis. Make a one-sentence comment about your plot.

```python
newdf['pclass'].value_counts().plot(kind='bar')
plt.xlabel("Class of Passengers")
plt.ylabel("Count of Passengers in each class")
plt.title("Count of Passengers by Class")
```

Number of People in the first, second and third class of the titanic ship

```python
newdf['embarked'].value_counts().plot(kind='barh')
plt.xlabel("Count of Passengers embarking")
plt.ylabel("Location of Embarking")
plt.title("Count of Passengers embarking the journey from 3 departure points")
```

Number of people who have boarded the ship from Cherbourg, Queenstown and Southampton

### Problem 3 (Pie chart)
a) Do problem 2(a) with a pie chart. Also, plot the % numbers in each piece of the pie.

b) Do problem 2(b) with a pie chart. Also, plot the % numbers in each piece of the pie.

```python
newdf['pclass'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.axis('equal')
plt.title("Ticket class of Passengers")
plt.legend(['Third Class', 'First Class', 'Second Class'], loc='center right')
```

Pie Chart Denoting Number of People in the first, second and third class of the titanic ship

```python
newdf['embarked'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.axis('equal')
plt.title("Boarding Loactions of Passengers")
plt.legend(['Southampton', 'Cherbourg', 'Queenstown'], loc='lower right')
```

Pie chart denoting number of people who have boarded the ship from Cherbourg, Queenstown and Southampton

### Problem 4 (Histogram)
a) Draw histogram of the age variable from the original data with ten bins. Explain the graph. Some hints are here. https://mode.com/example-gallery/python_histogram/

b) Draw a histogram of the variable fare from the cleaned dataset. Try a various number of bins and see how the histogram looks like. Choose a number of bins that you think is appropriate. Explain the process.

```python
df.hist(column='age', bins=10, grid=False, figsize=(10,10))
plt.xlabel("Age of People Boarding")
plt.ylabel("Number of passengers")
plt.title("Count of the age of Passengers boarding Titanic")
```

This histogram visualises the number of people in the given age frame of 0-80 years who boarded the ship Titanic. It has 10 bins specified.

```python
newdf.hist(column='fare', bins=15, grid=False, figsize=(8,10))
plt.xlabel("Number of passengers")
plt.ylabel("Fare Paid")
plt.title("Fare paid by passengers who boarded the titanic")
```

I have chosen 15 bins as it can fit all data and can show the histograms clearly for the same. This histogram displayes fare paid by number of people who boarded the ship on the x axis.

### Problem 5 (Scatter plot and line graph)
a) Create a scatter plot of age (in x-axis) and fare (in the y-axis) and use the categorize function to color survived variable differently for different values.

b) Use the age variable again to draw a KDE line graph. Find what a KDE graph and explain it it is showing here.

```python
df.plot.scatter(x='age', y='fare', c='survived', cmap='seismic')
plt.xlabel("Age of Passengers")
plt.ylabel("Fare Paid")
plt.title("Scatterplot denoting age of passengers and the fare paid by them.")
```

```python
df['age'].plot.kde()

plt.title("Kernel Density Estimate for Age of Passengers")
```

In statistics, kernel density estimation (KDE) is a non-parametric way to estimate the probability density function (PDF) of a random variable. This function uses Gaussian kernels and includes automatic bandwidth determination. It is showing the probability density function of age of passengers who boarded the Titanic in this visualisation
