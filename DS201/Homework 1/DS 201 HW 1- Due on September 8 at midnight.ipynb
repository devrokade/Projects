### Problem 9. 
(a) Create a new column called "Class" with the following rules: A country with GDP $\leq$ 15000 - poor, A country with GDP between 25000 to 35000 - middle and the rest of the countries - rich. (b) Find out how many countries are in each class.

```python
nd= newdf #renamed data with a short name for simplicity
nd=newdf.reset_index(inplace=False) # changed the index to a column
nd.rename(columns={'index':'id'}, inplace=True) # remaed the index column as id 
```

```python
# creating a list of ids that will have unlucky label
ui= list(nd[(nd["GDP"]<=15000)].id)
# creating a list of ids that will have lucky label
li= list(nd[(nd["GDP"]>15000)& ((nd["GDP"]<=25000))].id)
## creating a list of ids that will have averageluck label
rest=list(set(nd.id)-(set(ui) | set(li)))
```

```python
#creating a column with .loc function
nd.loc[nd.id.isin(ui), 'Class'] = 'Poor'
nd.loc[nd.id.isin(li), 'Class'] = 'Middle'
nd.loc[nd.id.isin(rest), 'Class'] = 'Rich'
#renaming the data to titanic2
newdf = nd
newdf.head(5)
```

```python
nd.groupby(['Class']).size()
```
